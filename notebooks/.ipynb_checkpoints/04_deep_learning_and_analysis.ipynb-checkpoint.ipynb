```python
# === Cell 1: Imports, paths, and CSV adapter ===
from pathlib import Path
import os, json, random, numpy as np, pandas as pd

import torch
import matplotlib.pyplot as plt

# Repro
SEED = 42
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

# Your existing project structure
PROJ_ROOT = Path.home() / "Computer-Vision-Assignment"
RES_DIR   = PROJ_ROOT / "results"
SUB_DIR   = RES_DIR / "subset"
MET_DIR   = RES_DIR / "metrics"
FIG_DIR   = RES_DIR / "figures"
for d in [RES_DIR, SUB_DIR, MET_DIR, FIG_DIR]: d.mkdir(parents=True, exist_ok=True)

CSV_PATH = SUB_DIR / "subset_split.csv"
assert CSV_PATH.exists(), f"Run Phase 1 first; missing {CSV_PATH}"

# --- CSV adapter: add the columns your friend's code expects ---
df = pd.read_csv(CSV_PATH).copy()

# If you already have class names as strings in "label", keep them; else treat "label" as class name.
# Make a clean string column "class_name"
df["class_name"] = df["label"].astype(str)

# Absolute path column
df["image_path_abs"] = df["image_path"].apply(lambda p: str(Path(p).resolve()))

# Save back so the rest of the notebook can use it
df.to_csv(CSV_PATH, index=False)
print("CSV adapted with columns:", {"class_name", "image_path_abs"})
print("Rows:", len(df["image_path_abs"].unique()), "images across", df["split"].unique())
```

```python
# === Cell 2: Dataset + DataLoaders (with normalization & optional class-balance) ===
from PIL import Image
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler
import torchvision.transforms as T
from collections import Counter

df = pd.read_csv(CSV_PATH)
classes = sorted(df["class_name"].unique())
cls2idx = {c:i for i,c in enumerate(classes)}
idx2cls = {i:c for c,i in cls2idx.items()}

class TrafficSignDataset(Dataset):
    def __init__(self, csv_path, split="train", transform=None):
        d = pd.read_csv(csv_path)
        d = d[d["split"].str.lower()==split].copy()
        self.paths  = d["image_path_abs"].tolist()
        self.labels = d["class_name"].map(cls2idx).astype(int).tolist()
        self.transform = transform
    def __len__(self): return len(self.paths)
    def __getitem__(self, i):
        img = Image.open(self.paths[i]).convert("RGB")
        if self.transform: img = self.transform(img)
        return img, torch.tensor(self.labels[i], dtype=torch.long)

train_tf = T.Compose([
    T.Resize((128,128)),
    T.RandomHorizontalFlip(),
    T.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
    T.ToTensor(),
    T.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225]),
])

val_tf = T.Compose([
    T.Resize((128,128)),
    T.ToTensor(),
    T.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225]),
])

train_ds = TrafficSignDataset(CSV_PATH, "train", train_tf)
val_ds   = TrafficSignDataset(CSV_PATH, "val",   val_tf)
test_ds  = TrafficSignDataset(CSV_PATH, "test",  val_tf)

# Optional: weighted sampling for imbalance
labels = [train_ds[i][1].item() for i in range(len(train_ds))]
cnt = Counter(labels)
num_classes = len(classes)
class_count = np.array([cnt.get(i,0) for i in range(num_classes)], dtype=np.float32)
class_count = np.maximum(class_count, 1.0)
class_weight = 1.0 / class_count
sample_weight = np.array([class_weight[y] for y in labels], dtype=np.float32)
sampler = WeightedRandomSampler(torch.from_numpy(sample_weight), num_samples=len(sample_weight), replacement=True)

train_loader = DataLoader(train_ds, batch_size=32, sampler=sampler, shuffle=False, num_workers=2)
val_loader   = DataLoader(val_ds,   batch_size=32, shuffle=False, num_workers=2)
test_loader  = DataLoader(test_ds,  batch_size=32, shuffle=False, num_workers=2)

xb, yb = next(iter(train_loader))
print("Sanity batch:", xb.shape, yb[:8].tolist())
print("Classes:", classes)
```

```python
# === Cell 3: CNN model ===
import torch.nn as nn
import torch.optim as optim

class SimpleCNN(nn.Module):
    def __init__(self, num_classes):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3,32,3,1,1), nn.BatchNorm2d(32), nn.ReLU(), nn.MaxPool2d(2),
            nn.Conv2d(32,64,3,1,1), nn.BatchNorm2d(64), nn.ReLU(), nn.MaxPool2d(2),
            nn.Conv2d(64,128,3,1,1), nn.BatchNorm2d(128), nn.ReLU(), nn.MaxPool2d(2),
        )
        self.classifier = nn.Sequential(
            nn.Dropout(0.5),
            nn.Linear(128*16*16, 256), nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, num_classes)
        )
    def forward(self,x):
        x = self.features(x)
        x = x.view(x.size(0), -1)
        return self.classifier(x)

model = SimpleCNN(num_classes=len(classes)).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3, weight_decay=1e-5)
scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=5, gamma=0.5)
print(model)
```

```python
# === Cell 4: Train loop ===
EPOCHS = 10
train_losses, val_losses, val_accs = [], [], []

for epoch in range(1, EPOCHS+1):
    model.train(); run_loss = 0.0
    for xb, yb in train_loader:
        xb, yb = xb.to(device), yb.to(device)
        optimizer.zero_grad()
        out = model(xb)
        loss = criterion(out, yb)
        loss.backward()
        optimizer.step()
        run_loss += loss.item() * xb.size(0)
    scheduler.step()
    tr_loss = run_loss / len(train_loader.dataset)
    train_losses.append(tr_loss)

    # val
    model.eval(); correct=0; tot=0; vloss=0.0
    with torch.no_grad():
        for xb, yb in val_loader:
            xb, yb = xb.to(device), yb.to(device)
            out = model(xb)
            vloss += criterion(out, yb).item() * xb.size(0)
            pred = out.argmax(1)
            correct += (pred==yb).sum().item()
            tot += yb.size(0)
    val_losses.append(vloss/len(val_loader.dataset))
    val_accs.append(correct/tot)
    print(f"Epoch[{epoch}/{EPOCHS}] tr_loss={tr_loss:.4f} val_loss={val_losses[-1]:.4f} val_acc={val_accs[-1]:.4f}")

(PROJ_ROOT / "results" / "models").mkdir(parents=True, exist_ok=True)
torch.save(model.state_dict(), PROJ_ROOT / "results" / "models" / "cnn_simple.pth")

# curves
plt.figure(figsize=(8,3.5))
plt.plot(train_losses, label="train")
plt.plot(val_losses,   label="val")
plt.title("CNN Loss")
plt.xlabel("epoch"); plt.ylabel("loss"); plt.legend(); plt.tight_layout()
plt.savefig(FIG_DIR/"cnn_loss.png", dpi=160); plt.show()

plt.figure(figsize=(6,3.5))
plt.plot(val_accs, label="val acc")
plt.title("CNN Val Accuracy"); plt.xlabel("epoch"); plt.ylabel("acc"); plt.legend(); plt.tight_layout()
plt.savefig(FIG_DIR/"cnn_val_acc.png", dpi=160); plt.show()
```

```python
# === Cell 5: Test evaluation + confusion matrix ===
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report

model.eval()
all_y, all_p = [], []
with torch.no_grad():
    for xb, yb in test_loader:
        xb = xb.to(device)
        logits = model(xb)
        preds = logits.argmax(1).cpu().numpy()
        all_p.extend(preds)
        all_y.extend(yb.numpy())

acc = (np.array(all_y)==np.array(all_p)).mean()
print(f"Test Accuracy: {acc:.4f}")

cm = confusion_matrix(all_y, all_p, labels=list(range(len(classes))), normalize="true")
plt.figure(figsize=(7,6))
sns.heatmap(cm, annot=True, fmt=".2f", cmap="Blues",
            xticklabels=classes, yticklabels=classes)
plt.title("CNN Confusion Matrix  TEST")
plt.xlabel("Pred"); plt.ylabel("True"); plt.tight_layout()
plt.savefig(FIG_DIR/"cnn_confmat_test.png", dpi=160); plt.show()

# Detailed per-class report (CSV for bonus)
rep = classification_report(all_y, all_p, target_names=classes, output_dict=True, zero_division=0)
pd.DataFrame(rep).T.to_csv(MET_DIR/"cnn_report.csv")
print("Saved:", MET_DIR/"cnn_report.csv")
```

```python
# === Cell 6: Phase-4 comparison + failure images + insights ===
import json, cv2
from sklearn.metrics import accuracy_score

# 1) Bring classical ML summary if present (your Phase-2 saved model_summary.json)
phase2_summary = (MET_DIR/"model_summary.json")
df_ml = None
if phase2_summary.exists():
    with open(phase2_summary) as f: ms = json.load(f)
    df_ml = pd.DataFrame(ms).T.reset_index(names="model")
    df_ml.rename(columns={"test_acc":"test_accuracy","val_acc":"cv_mean"}, inplace=True)

# 2) CNN row
cnn_row = pd.DataFrame([{"model":"CNN","test_accuracy":float(acc),"cv_mean":float(np.mean(val_accs)),"cv_std":float(np.std(val_accs))}])

# Combine
df_all = cnn_row if df_ml is None else pd.concat([df_ml[["model","test_accuracy","cv_mean"]], cnn_row], ignore_index=True)
display(df_all)

plt.figure(figsize=(5,4))
plt.bar(df_all["model"], df_all["test_accuracy"])
plt.ylim(0,1); plt.title("Model Accuracy Comparison (Phase 4)")
plt.ylabel("Test Accuracy"); plt.tight_layout()
plt.savefig(FIG_DIR/"phase4_model_comparison.png", dpi=160); plt.show()

# 3) Save top-5 failure images for report
fail_idx = np.where(np.array(all_y)!=np.array(all_p))[0]
print("Misclassified:", len(fail_idx))

# build crops quickly (same util you used before)
mini = df[df["split"].str.lower()=="test"].reset_index(drop=True)
def load_crop_bgr(row, pad=4):
    p = row["image_path_abs"]
    x,y,w,h = int(row["x"]), int(row["y"]), int(row["w"]), int(row["h"])
    bgr = cv2.imread(p); H,W = bgr.shape[:2]
    x0,y0 = max(0,x-pad), max(0,y-pad)
    x1,y1 = min(W,x+w+pad), min(H,y+h+pad)
    crop = bgr[y0:y1, x0:x1]
    return crop if crop.size>0 else bgr

for k, idx in enumerate(fail_idx[:5]):
    row = mini.iloc[idx]
    crop = load_crop_bgr(row)
    rgb = cv2.cvtColor(crop, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(4,3)); plt.imshow(rgb); plt.axis("off")
    plt.title(f"True: {classes[all_y[idx]]}\nPred: {classes[all_p[idx]]}", fontsize=9)
    plt.tight_layout(); out = FIG_DIR/f"phase4_fail_cnn_{k+1}.png"
    plt.savefig(out, dpi=160); plt.show()
    print("Saved:", out)

# 4) Short insights note
txt = f"""Phase 4  Comparative Notes
* Classical (Phase 2) vs CNN: see phase4_model_comparison.png.
* CNN val acc mean  {np.mean(val_accs):.3f}, test acc  {acc:.3f}.
* Likely issues: limited data and class imbalance  the other-sign dominates confusions.
* Next: stronger aug, more epochs, or fine-tune a pretrained ResNet-18 for better features.
"""
(MET_DIR/"phase4_text_insights.md").write_text(txt); print(txt)
```
